```python
from datetime import datetime
from SPARQLWrapper import SPARQLWrapper, JSON, XML, N3, RDF
import sys
```

```python
def get_results(endpoint_url, query):
    user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
    # TODO adjust user agent; see https://w.wiki/CX6
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()

endpoint_url = 'http://localhost:3030/Satadet/query'

query = """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
SELECT ?subject ?predicate ?object
WHERE {
  ?subject ?predicate ?object
}
"""
results1 = get_results(endpoint_url, query)
```

```python
print('<html><head><title>Our Tree Ontology</title></head>')

print('<body><h1>Ontologies of the Trees</h1>') 

print('<ul>')

for result in results1["results"]["bindings"]:
    if("subject" in result):
        gelenler = result["subject"]["value"][73:]
        if len(gelenler) != 0:
            subject = gelenler
        else:
            subject = "NONE"
    else:
        subject = "NONE"
        
    if("object" in result):
        gelenler = result["object"]["value"][73:]
        if len(gelenler) != 0:
            object2 = gelenler
        else:
            object2 = "NONE"
    else:
        object2 = "NONE"

    if (subject != "NONE" and object2 != "NONE"):
        print('<li> -- {},{} </li>'.format(subject,object2))

print('</ul>')
print('</body></html>')
```

```python
endpoint_url2 = "https://query.wikidata.org/sparql"

query2 = """#Cats
SELECT ?item ?itemLabel ?thumbnail ?countryLabel ?coordinate
WHERE 
{
  ?item wdt:P31 wd:Q10884 . # Must be of a cat
  ?item wdt:P17 ?country.
  ?item wdt:P625 ?coordinate.
  OPTIONAL { ?item wdt:P18 ?thumbnail . }
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". } # Helps get the label in your language, if not, then en language
}"""
```

```python
results2 = get_results(endpoint_url2, query2)
```

```python
print('<html><head><title>Tree Ontologies</title></head>')

print('<body><h1>Ontologies of the Trees</h1>') 

print('<ul>')

for result in results2["results"]["bindings"]:
    if("itemLabel" in result):
        itemLabel = result["itemLabel"]["value"]
    else:
        itemLabel = "NONE"
    
    if("thumbnail" in result):
        thumbnail = result["thumbnail"]["value"]
    else:
        thumbnail = "NONE"
    
    if("countryLabel" in result):
        countryLabel = result["countryLabel"]["value"]
    else:
        countryLabel = "NONE"
    
    if("coordinate" in result):
        coordinate = result["coordinate"]["value"]
    else:
        coordinate = "NONE"


    print('<li> -- <img src="{}" height="60px">, {},{},{} </li>'.format(thumbnail.replace("300","60"),itemLabel,countryLabel,coordinate))

print('</ul>')
print('</body></html>')
```
